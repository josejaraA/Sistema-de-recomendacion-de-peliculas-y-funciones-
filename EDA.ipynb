```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
movies_df = pd.read_csv('Dataset_LimpioV.csv')
movies_df.head()
```

```python
movies_df.info()
```

```python
### calcular la media y rellenar los datos 
media_budget = movies_df[movies_df['budget'] != 0]['budget'].mean()
media_votes = movies_df[movies_df['vote_average'] !=0]['vote_average'].mean()
media_popularity = movies_df[movies_df['popularity'] !=0]['popularity'].mean()
### Reemplazar los valores que son 0 con la media calculada
movies_df['budget'] = movies_df['budget'].replace(0, media_budget)
movies_df['vote_average'] = movies_df['vote_average'].replace(0, media_votes)
movies_df['popularity'] = movies_df['popularity'].replace(0, media_popularity)
```

```python
movies_df.describe()
```

```python
### Anlisis de relaciones

# Crear un scatter plot para analizar la relacin entre presupuesto y ingresos
plt.figure(figsize=(10, 6))
sns.scatterplot(x=movies_df['budget'], y=movies_df['revenue'])
plt.title('Relacin entre Presupuesto e Ingresos')
plt.xlabel('Presupuesto')
plt.ylabel('Ingresos')
plt.show()

# Crear un scatter plot para analizar la relacin entre popularidad y puntuacin promedio
plt.figure(figsize=(10, 6))
sns.scatterplot(x=movies_df['popularity'], y=movies_df['vote_average'])
plt.title('Relacin entre Popularidad y Puntuacin Promedio')
plt.xlabel('Popularidad')
plt.ylabel('Puntuacin Promedio')
plt.show()

# Crear un scatter plot para analizar la relacin entre duracin y popularidad
plt.figure(figsize=(10, 6))
sns.scatterplot(x=movies_df['runtime'], y=movies_df['popularity'])
plt.title('Relacin entre Duracin y Popularidad')
plt.xlabel('Duracin (minutos)')
plt.ylabel('Popularidad')
plt.show()
```

```python
# Separar los gneros en mltiples filas
df_genres = movies_df['genres'].str.split(' ', expand=True).stack().reset_index(level=1, drop=True)
df_genres.name = 'genre'

# Calcular la frecuencia de cada gnero
genre_counts = df_genres.value_counts()

# Visualizacin de Frecuencia de Gneros
plt.figure(figsize=(10, 6))
sns.barplot(x=genre_counts.index, y=genre_counts.values)
plt.title('Frecuencia de Gneros')
plt.xlabel('Gneros')
plt.ylabel('Frecuencia')
plt.xticks(rotation=45)
plt.show()
```
