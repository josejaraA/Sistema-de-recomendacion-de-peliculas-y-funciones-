1. Limpieza y transformación del dataset movies_datset.csv

```python
import pandas as pd
import ast
import json
peliculas = pd.read_csv('movies_dataset.csv')
peliculas.head()
```

```python
###Eliminacion de columnas innecesrias 

eliminar =['adult','belongs_to_collection','imdb_id','original_title','poster_path','homepage','production_countries','status','video']

peliculasdata = peliculas.drop(columns = eliminar )

peliculasdata.to_csv('movies_dataset_filtrados.csv',index=False)
```

```python
 ### conversión de release date a datetime

movies_df = pd.read_csv('movies_dataset_filtrados.csv')

movies_df['release_date'] = pd.to_datetime(movies_df['release_date'], errors='coerce')
### Conversión de id a un entero natural de python
peliculas = peliculas[peliculas['id'].apply(lambda x: str(x).isdigit())]
peliculas['id'] = peliculas['id'].astype(int)
```

```python
movies_df.head()
```

```python
### nueva columna para el mes de estreno
meses_espanol = ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo', 'Junio', 'Julio', 'Agosto', 'Septiembre', 'Octubre', 'Noviembre', 'Diciembre']

movies_df['release month'] = movies_df['release_date'].apply(lambda x:meses_espanol[x.month-1] if pd. notnull(x) else None)
```

```python
### nueva columna para el dia de estreno 

dias_espanol = ['Lunes','Martes','Miércoles','Jueves','Viernes','Sábadado','Domingo']

movies_df['release_day'] = movies_df['release_date'].apply(lambda x:dias_espanol[x.weekday()] if pd.notnull(x) else None)
```

```python
### desanidacion y conversion de los generos en una cadena
def desanidar_generos(genres_str):
    genres_list = ast.literal_eval(genres_str)
    return ' '.join([genre['name'] for genre in genres_list])
movies_df['genres'] = movies_df['genres'].apply(desanidar_generos)
```

```python
### Guardamos el dataset en un nuevo archivo
movies_df.to_csv('Data_transformada.csv', index=False)
```

```python
movies_trans = pd.read_csv('Data_transformada.csv')
movies_trans.head()
```

2. Limpieza y transformación del dataset credits_csv

```python
### Cargar el archivo CSV en un DataFrame
credits_df = pd.read_csv('credits.csv')  

credits_df.head()
```

```python
### Extracción del nombre de los actores del campo cast
def obtener_nombres_actores(cadena_cast):
    try:
        cast_list = json.loads(cadena_cast.replace("'", "\""))
        return [actor['name'] for actor in cast_list]
    except json.JSONDecodeError:
        return []

# Aplicar la función para obtener una columna con los nombres de los actores
credits_df['actores'] = credits_df['cast'].apply(obtener_nombres_actores)

### Extracción del nombre de los directores del campo crew
def obtener_nombres_directores(cadena_crew):
    try:
        crew_list = json.loads(cadena_crew.replace("'", "\""))
        return [persona['name'] for persona in crew_list if persona['job'] == 'Director']
    except json.JSONDecodeError:
        return []
credits_df['Directores'] = credits_df['crew'].apply(obtener_nombres_directores)
```

```python
eliminar2 = ['cast', 'crew']

credits_df = credits_df.drop(columns=eliminar2)
```

```python
### Guardamos el dataset en un nuevo archivo
credits_df.to_csv('credits_limpio.csv',index=False)
```

```python
###Unificamos los datasets por el id y lo guardamos 

union = pd.merge(movies_trans, credits_df, on='id', how='inner')
union.to_csv('Dataset_LimpioV.csv')
```
